```python
from ultralytics import YOLO
import torch
import cv2

model = YOLO("yolov8n.pt")

model.train(
    data="/kaggle/input/datasets/beyzakucuk/ppe-detection-v1/data.yaml",
    epochs=50,
    imgsz=512,        
    batch=16,         
    workers=2,        
    device=0,
    half=True,        
    cache=True,      
    patience=5
)
```

```python

from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/detect/train2/weights/best.pt")

metrics = model.val()
print(metrics)
```
